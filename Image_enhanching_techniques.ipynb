```python
from PIL import Image as im
import numpy as np
```

```python
path = '/content/test4.jpg'
```

```python
img  = im.open(path).convert('L')
imgarray = np.array(img)
size = imgarray.shape
```

```python
threshold = 50
```

# **Thresholding**

---



```python
wholeimg = []
wr =0
for i in range(size[0]):
  rows = []
  for j in range(size[1]):
    if imgarray[i][j]<threshold:
      wr = 0
    else:
      wr = 255
    rows.append(wr)
  wholeimg.append(rows)
csr=np.array(wholeimg)
csr=np.uint8(csr)
imr  = im.fromarray(csr)
```

```python
imr
```

# **Grey level sclicing without background**

---



```python
r1 = 70
r2 = 230
s1 = 0
wholeimg = []
wr =0
for i in range(size[0]):
  rows = []
  for j in range(size[1]):
    if imgarray[i][j]<r1:
      wr = s1
    elif imgarray[i][j]>r2:
      wr = s1
    else:
      wr = 255
    rows.append(wr)
  wholeimg.append(rows)
csr=np.array(wholeimg)
csr=np.uint8(csr)
imr  = im.fromarray(csr)
```

```python
imr
```

# **Grey level sclicing with background**

---



```python
r1 = 190
r2 = 210
s1 = 0
wholeimg = []
wr =0
for i in range(size[0]):
  rows = []
  for j in range(size[1]):
    if imgarray[i][j]<r1:
      wr = imgarray[i][j]
    elif imgarray[i][j]>r2:
      wr = imgarray[i][j]
    else:
      wr = 255
    rows.append(wr)
  wholeimg.append(rows)
csr=np.array(wholeimg)
csr=np.uint8(csr)
imr  = im.fromarray(csr)
```

```python
imr
```

# **Contras Streching**

---



```python
r1 = 80   #note: s1<r1 and s2>r2
r2 = 190
s1 = 50
s2 = 210

l = s1/r1
m = (s2-s1)/(r2-r1)
n = (255-s1)/(255-r2)

wholeimg = []
wr =0
for i in range(size[0]):
  rows = []
  for j in range(size[1]):
    if imgarray[i][j]<r1:
      wr = l*imgarray[i][j]
    elif imgarray[i][j]>r2:
      wr = (n*(imgarray[i][j]-r2))+s2
      if wr>255:
        wr=255
    else:
      wr = (m*(imgarray[i][j]-r1))+s1
      if wr>255:
        wr=255
    rows.append(wr)
  wholeimg.append(rows)
csr=np.array(wholeimg)
csr=np.uint8(csr)
imr  = im.fromarray(csr)
```

```python
imr
```
