```python
from PIL import Image as im
```

```python
path = '/content/test4.jpg'
```

```python
import numpy as np
```

```python
def shrink(path):
  img  = im.open(path).convert('L')
  imgarray = np.array(img)
  si = imgarray.shape
  wholeimg = []
  for i in range(0,si[0],2):
    rows = []
    for j in range(0,si[1],2):
      rows.append(imgarray[i][j])
    wholeimg.append(rows)
  csr=np.array(wholeimg)
  imr  = im.fromarray(csr)
  return imr
```

```python
def zoom(path):
  img  = im.open(path).convert('L')
  imgarray = np.array(img)
  si = imgarray.shape
  wholeimg = np.zeros((si[0]*2,si[1]*2),int)
  p = 0
  for i in range(0,si[0]*2,2):
    rows = []
    q =0
    for j in range(0,si[1]*2,2):
      wholeimg[i][j] = imgarray[p][q]
      q+=1
    p+=1
  for i in range(2,si[0]*2,2):
    for j in range(0,si[1]*2,2):
      wholeimg[i-1][j] = round((wholeimg[i-2][j]+wholeimg[i][j])/2)
  for i in range(2,si[0]*2):
    for j in range(2,si[1]*2,2):
      wholeimg[i][j-1] = round((wholeimg[i][j]+wholeimg[i][j-2])/2)
  csr=np.array(wholeimg)
  # for i in range(0,si[0]*2):
  #   for j in range(0,si[1]*2):
  #     print(wholeimg[i][j],end=" ")
  #   print()
  csr=np.uint8(csr)
  imr  = im.fromarray(csr)
  return imr
```
